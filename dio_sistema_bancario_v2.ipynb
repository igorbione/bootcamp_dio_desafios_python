## Desafio de projeto - Criando um sistema bancário com Python com funções

```python
def criarUsuario():

    nome = input("Digite seu nome: ")
    data_nascimento = input("Digite sua data de nascimento: ")
    endereco = input("Digite seu endereço: ")
    cpf = input("Digite seu CPF: ")

    if cpf in usuarios:
        print("ERRO: CPF já cadastrado.")
        return
    

    
    usuarios[cpf] = {
        'nome': nome,
        'data_nascimento': data_nascimento,
        'endereco': endereco,
        'contas': []
    }
    print("Usuário cadastrado com sucesso.")


usuarios = {} 

def criarConta(usuarios, cpf):
    if cpf not in usuarios:
        print("ERRO: CPF não cadastrado.")
        return
    
    usuario = usuarios[cpf]
    numero_conta = len(usuario['contas']) + 1
    agencia = '001'  # Fixed agency number for simplicity

    nova_conta = {
        'agencia': agencia,
        'numero': numero_conta,
        'saldo': 0
    }
    usuario['contas'].append(nova_conta)
    print(f"Conta criada com sucesso!\nNúmero da conta: {numero_conta}\nAgência: {agencia}")


def deposito(saldo: float, extrato_historico: list, valor_depo: int):
    if valor_depo > 0:
        saldo += valor_depo
        print("Valor depositado com sucesso!")
        print("Novo saldo: {}".format(saldo))

        extrato_historico.append(f"Depósito: R$ {valor_depo:.2f}")
    else:
        print("ERRO: Só podem ser adicionados valores maiores que zero!")
    return saldo

def saque(saldo: int, extrato_historico: list, n_saque: int, limite_saques: int, limite: float, valor_saque: float):
    condicao_limite = valor_saque > saldo
    if condicao_limite:
        print("ERRO: Você não tem saldo!")

    condicao_saque = valor_saque > limite
    if condicao_saque:
        print("ERRO: Excedeu o limite de saque por transação de {}!".format(limite))

    condicao_n_saques = n_saque >= limite_saques
    if condicao_n_saques:
        print("ERRO: Excedeu o limite de saque diário de {}!".format(limite_saques))

    if not condicao_limite and not condicao_saque and not condicao_n_saques:
        saldo -= valor_saque
        print("Valor sacado com sucesso!")
        n_saque += 1
        extrato_historico.append(f"Saque: R$ {valor_saque:.2f}")
    return saldo, n_saque

def extrato(saldo: float, extrato_historico: list):
    print("###### Extrato ########")
    if not extrato_historico:
        print("Não foram realizadas transações")
    else:
        for transacao in extrato_historico:
            print(transacao)
    print("Saldo atual: R$ {:.2f}".format(saldo))



criarUsuario()

menu = """
[a] # Criar usuario
[cc] # Criar conta
[d] # Deposito
[s] # Saque
[e] # Extrato
[q] # Encerrar atendimento

=> """

saldo = 0.0
limite = 500
extrato_historico = []
N_SAQUE = 0
LIMITE_SAQUES = 3


while True:

    operacao = input(menu)

    if operacao == "a":
        criarUsuario()

    elif operacao == "cc":
        criarConta(usuarios,input("Digite seu CPF: ") )
    elif operacao == "d":
        valor_depo = int(input("Digite o valor:"))
        deposito(saldo, extrato_historico,valor_depo)
        
       
    elif operacao == "s":
        valor_saque = int(input("Digite o valor a ser sacado:"))
        saque(saldo, extrato_historico,N_SAQUE, LIMITE_SAQUES, limite, valor_saque)
       
    elif operacao == "e":
        extrato(saldo,extrato_historico)
     
    elif operacao == "q":
        break
    else:
        print("Operação inválida: Seleciona d para Deposito, s para Saque e para Extrato e q para Sair")
```
